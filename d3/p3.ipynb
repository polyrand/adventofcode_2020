```python
with open("input_p3.txt") as f:
    data = f.read().split("\n")
```

```python
total_rows = len(data)
```

```python
tree = "#"
empty = "."
```

```python
len_row = len(data[0])
```

```python
%%time
trees = 0

pattern = (3, 1)

current_pos = (0, 0)

while current_pos[1] < (total_rows - 1):
    new_pos = (current_pos[0] + pattern[0], current_pos[1] + pattern[1])

    if new_pos[0] >= len(data[0]):
        new_pos = (new_pos[0] - len(data[0]), new_pos[1])

    if data[new_pos[1]][new_pos[0]] == tree:
        trees += 1

    current_pos = new_pos


print(f"\nNumber of trees for pattern {pattern} = {trees}\n")
```

```python
%%time
total_trees = 1

patterns = [(3, 1), (1, 1), (5, 1), (7, 1), (1, 2)]


for pattern in patterns:

    current_pos = (0, 0)

    trees = 0

    while current_pos[1] < (total_rows - 2):
        new_pos = (current_pos[0] + pattern[0], current_pos[1] + pattern[1])

        if new_pos[0] >= len(data[0]):
            new_pos = (new_pos[0] - len(data[0]), new_pos[1])

        if data[new_pos[1]][new_pos[0]] == tree:
            trees += 1

        current_pos = new_pos

        if current_pos[1] >= total_rows - 1:
            break
            
    total_trees *= trees

print(f"\nNumber of trees for all patterns {patterns} = {total_trees}\n")
```
